```python
import gensim
import multiprocessing
from sklearn.metrics import mean_absolute_error, accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np 
np.set_printoptions(threshold=10000,suppress=True) 
import pandas as pd
import warnings
import matplotlib.pyplot as plt 
warnings.filterwarnings('ignore')
import gensim
```

```python
def train_model(corpus, model_size=100, save_path='word2vec_entraine.h5'):
    #corpus = corpus.apply(lambda line: gensim.utils.simple_preprocess(line))
    #corpus= gensim.utils.simple_preprocess(text) for text in corpus
    corpus = [gensim.utils.simple_preprocess(text) for text in corpus]
    cores = multiprocessing.cpu_count()

    model = gensim.models.Word2Vec(
        corpus,
        vector_size=model_size,
        sg=0,
        window=5,
        min_count=2,
        workers=cores-1
    )

    for i in range(model_size):
        model.train(corpus, total_examples=len(corpus), epochs=1)
        print(i, end=' ')

    model.save(save_path)
    return model
```

```python
data=pd.read_csv('../data/PubMed-multi-label-dataset.csv')
```

```python
corpus=data.iloc[:,2].values
```

```python
model=train_model(corpus, model_size=200, save_path='modele/word2vec_entraine.h5')
```
